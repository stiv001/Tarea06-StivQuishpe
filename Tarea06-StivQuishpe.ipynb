Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie y el Polinomio de Lagrange:

1. $\frac{1}{25x^{2}+1}$, $x_0 = 0$

Usando la Serie de Taylor:

```python
import numpy as np
import math
import sympy as sp
import matplotlib.pyplot as plt

# Definir la variable simbólica y la función
x = sp.Symbol('x')
f = 1 / (25 * (x**2) + 1)

# Función recursiva para calcular derivadas
def derivada_recursiva(funcion, n):
    if n == 0:
        return funcion
    else:
        return sp.diff(derivada_recursiva(funcion, n - 1), x)

# Función para calcular la serie de Taylor
def serie_taylor(funcion, variable, num_terminos, punto_inicial=0):
    taylor_exp = 0
    for n in range(num_terminos):
        derivada = derivada_recursiva(funcion, n)
        derivada_en_punto = derivada.subs(variable, punto_inicial)
        taylor_exp += (derivada_en_punto / math.factorial(n)) * (variable - punto_inicial)**n
    return taylor_exp

# Valores de x para la gráfica
x_vals = np.linspace(-1, 1, 100)
f_lambdified = sp.lambdify(x, f, 'numpy')
y_vals = f_lambdified(x_vals)

# Graficar la función original
plt.plot(x_vals, y_vals, label="Función original", color="blue", linewidth=2)

print("Fórmulas y Valores Simplificados de las Aproximaciones de Taylor")
colors = ['red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']
for n in range(1, 10):
    taylor_exp = serie_taylor(f, x, n)
    taylor_lambdified = sp.lambdify(x, taylor_exp, 'numpy')

    try:
        y_taylor_vals = taylor_lambdified(x_vals)
        if len(y_taylor_vals) == len(x_vals):
            plt.plot(x_vals, y_taylor_vals, linestyle="--", label=f"Aprox. Taylor orden {n}", color=colors[n-1])
    except Exception as e:
        print(f"Error evaluando el polinomio de Taylor de orden {n}")
        continue
        
    print(f"Orden {n}:")
    print(f"  Fórmula: {sp.simplify(taylor_exp)}")

# Configuración de la gráfica
plt.title("Aproximaciones de Taylor")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid(True)
plt.ylim(-1, 2)
plt.show()
```

Polinomio de Lagrange

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la función para calcular el polinomio de Lagrange
def polinomio_lagrange(puntos_x, puntos_y):
    x = sp.Symbol('x')
    n = len(puntos_x)
    P = 0
    for k in range(n):
        Lk = 1
        for i in range(n):
            if i != k:
                Lk *= (x - puntos_x[i]) / (puntos_x[k] - puntos_x[i])
        P += puntos_y[k] * Lk
    return sp.simplify(P)

# Puntos de datos
puntos_x = [-0.15, 0, 0.15]
puntos_y = [
    (1 / (25 * ((-0.15)**2) + 1)),
    (1 / (25 * ((0)**2) + 1)),
    (1 / (25 * (0.15**2) + 1))
]

# Calcular el polinomio de Lagrange
P = polinomio_lagrange(puntos_x, puntos_y)
print(f"Polinomio de Lagrange: {P}")

# Convertir el polinomio a una función utilizable con numpy
P_lambdified = sp.lambdify(sp.Symbol('x'), P, 'numpy')

# Valores de x para la gráfica
valores_x = np.linspace(-0.2, 0.2, 100)
valores_y_original = 1 / (25 * (valores_x**2) + 1)
valores_y_lagrange = P_lambdified(valores_x)

# Graficar la función original y el polinomio de Lagrange
plt.figure(figsize=(4, 6))
plt.plot(valores_x, valores_y_original, label="Función original", color="green", linewidth=2)
plt.plot(valores_x, valores_y_lagrange, label="Polinomio de Lagrange", color="purple", linestyle="--")
plt.scatter(puntos_x, puntos_y, color="red", label="Puntos originales", zorder=5)
plt.title("Comparación: Función Original vs Aproximación de Lagrange")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid()
plt.show()
```

Polinomio de Lagrange

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la función para calcular el polinomio de Lagrange
def polinomio_lagrange(puntos_x, puntos_y):
    x = sp.Symbol('x')
    n = len(puntos_x)
    P = 0
    for k in range(n):
        Lk = 1
        for i in range(n):
            if i != k:
                Lk *= (x - puntos_x[i]) / (puntos_x[k] - puntos_x[i])
        P += puntos_y[k] * Lk
    return sp.simplify(P)

# Puntos de datos
puntos_x = [-0.15, 0, 0.15]
puntos_y = [
    np.arctan(-0.15),
    np.arctan(0),
    np.arctan(0.15)
]

# Calcular el polinomio de Lagrange
P = polinomio_lagrange(puntos_x, puntos_y)
print(f"Polinomio de Lagrange: {P}")

# Convertir el polinomio a una función utilizable con numpy
P_lambdified = sp.lambdify(sp.Symbol('x'), P, 'numpy')

# Función original lambdificada
f_original_lambdified = np.arctan

# Valores de x para la gráfica
valores_x = np.linspace(-2, 2, 100)
valores_y_original = f_original_lambdified(valores_x)
valores_y_lagrange = P_lambdified(valores_x)

# Graficar la función original y el polinomio de Lagrange
plt.figure(figsize=(4, 6))
plt.plot(valores_x, valores_y_original, label="Función original", color="green", linewidth=2)
plt.plot(valores_x, valores_y_lagrange, label="Polinomio de Lagrange", color="purple", linestyle="--")
plt.scatter(puntos_x, puntos_y, color="red", label="Puntos originales", zorder=5)
plt.title("Comparación: Función Original vs Aproximación de Lagrange")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid()
plt.show()
```
